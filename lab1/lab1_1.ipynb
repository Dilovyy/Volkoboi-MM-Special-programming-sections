```python
file_path = "household_power_consumption.txt"

#df = pd.read_csv( file_path, sep=";", decimal=",")
df = pd.read_csv( file_path, sep=";", decimal=",",parse_dates={"Datetime": ["Date", "Time"]})
df = df.dropna()
#df["Date"] = pd.to_datetime(df["Date"], format="%d/%m/%Y")
#df["Time"] = pd.to_datetime(df["Time"], format="%H:%M:%S").dt.strftime("%H:%M")
#for col in df.columns.drop(["Date", "Time"]):
 #   df[col] = df[col].str.replace(",", ".").astype(float)
for col in df.columns.drop("Datetime"):
    df[col] = df[col].str.replace(",", ".").astype(float)
display(df)
df.info()
```

Обрати всі записи, у яких загальна активна споживана потужність перевищує 5 кВт.

```python
def select_active_power_bigger_than5(df):
    return df[df["Global_active_power"] > 5]

display(select_active_power_bigger_than5(df))
```

Additional Variable Information

1.date: Date in format dd/mm/yyyy
2.time: time in format hh:mm:ss
3.global_active_power: household global minute-averaged active power (in kilowatt)
4.global_reactive_power: household global minute-averaged reactive power (in kilowatt)
5.voltage: minute-averaged voltage (in volt)
6.global_intensity: household global minute-averaged current intensity (in ampere)
7.sub_metering_1: energy sub-metering No. 1 (in watt-hour of active energy). It corresponds to the kitchen, containing mainly a dishwasher, an oven and a microwave (hot plates are not electric but gas powered).
8.sub_metering_2: energy sub-metering No. 2 (in watt-hour of active energy). It corresponds to the laundry room, containing a washing-machine, a tumble-drier, a refrigerator and a light.
9.sub_metering_3: energy sub-metering No. 3 (in watt-hour of active energy). It corresponds to an electric water-heater and an air-conditioner.

Обрати всі записи, у яких сила струму лежить в межах 19-20 А, для них виявити ті, у яких пральна машина та холодильних споживають більше, ніж бойлер та кондиціонер.

```python
def select_current_in_range19_20(df):
    subset = df[(df["Global_intensity"] >= 19) & (df["Global_intensity"] <= 20)]
    return subset[ (subset["Sub_metering_1"] + subset["Sub_metering_2"]) > (subset["Sub_metering_3"])]


select_current_in_range19_20(df)
```

Обрати випадковим чином 500000 записів (без повторів елементів вибірки), для них обчислити середні величини усіх 3-х груп споживання електричної енергії

```python
def random_sample_means(df):
    sample = df.sample(n=500_000, replace=False, random_state=4)
    #display(sample)
    return sample[["Sub_metering_1", "Sub_metering_2", "Sub_metering_3"]].mean()

random_sample_means(df)
```

Обрати ті записи, які після 18-00 споживають понад 6 кВт за хвилину в середньому, серед відібраних визначити ті, у яких основне споживання електроенергії у вказаний проміжок часу припадає на пральну машину, сушарку, холодильник та освітлення (група 2 є найбільшою), а потім обрати кожен третій результат із першої половини та кожен четвертий результат із другої половини.


```python
def last_vybirka(df):
    subset = df[(df["Datetime"].dt.hour >= 18) & (df["Global_active_power"]>6)]
    subset = subset[subset["Sub_metering_2"] > subset[["Sub_metering_1","Sub_metering_3"]].max(axis=1)]

    number = len(subset) // 2
    half1 = subset.iloc[:number:3]
    half2 = subset.iloc[number::4]
    return pd.concat([half1, half2])

last_vybirka(df)
```

```python
def correlation(df, col1, col2):
    pearson = df[col1].corr(df[col2], method="pearson").item()
    spearman = df[col1].corr(df[col2], method="spearman").item()
    return pearson, spearman

correlation(df, "Sub_metering_1", "Sub_metering_2")
```
